# 04-2 matplotlib 라이브러리 자유자재로 사용하기

```python
tips = sns.load_dataset("tips")
print(tips.head())
print(type(tips))
```

#### 히스토그램 그래프 그리기 (일변량 그래프, 변수 하나만 사용)

```python
fig = plt.figure()
axes1 = fig.add_subplot(1,1,1)
```

```python
# 히스토그램 생성, 열지정, x축간격 조정(bins)
axes1.hist(tips['total_bill'], bins=10)

#제목 설정
axes1.set_title('Histogram of Total Bill')

#x축 설정
axes1.set_xlabel('Frequency')

#y축 설정
axes1.set_ylabel('Total Bill')

fig
```

#### 산점도 그래프 그리기(이변량 그래프, 변수 2개를 사용)

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)

#산점도그래프의 x축, y축 지정
axes1.scatter(tips['total_bill'], tips['tip'])

#제목 지정
axes1.set_title('Scatterplot of Total Bill vs Tip')

#x축 지정
axes1.set_xlabel('Total Bill')

#y축 지정
axes1.set_ylabel('Tip')
```

#### 박스그래프 그리기 (이산형, 연속형 변수 함께 사용하는 그래프)

```python
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1,1,1)

axes1.boxplot([tips[tips['sex'] == 'Female']['tip'],  #성별이 Female인 데이터에서 tip 열 데이터 추출 리스트 전달
               tips[tips['sex'] == 'Male']['tip']],   #성별이 Male인 데이터에서 tip 열 데이터 추출 리스트 전달
               labels=['Female','Male'])

axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by Sex')
```

#### 다변량 그래프 그리기 - 산점도그래프
- 점의 색상을 다르게하거나, 점의 크기를 다르게 하는 방법으로 표현가능.

```python
# 성별(문자열)을 정수로 치환하는 함수
def recode_sex(sex):
    if sex == 'Female':
        return 0
    else:
        return 1
```

```python
print(tips)
```

```python
# apply 메소드를 사용하여 sex열에 recode_sex 함수를 브로드캐스팅.
tips['sex_color'] = tips['sex'].apply(recode_sex)
```

```python
print(tips)
```

```python
# 함수 생성
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
# 값 지정
axes1.scatter(
    x=tips['total_bill'], # x 축
    y=tips['tip'],        # y 축
    s=tips['size'] *10,       # s는 점의 크기를 의미
    c=tips['sex_color'],  # c는 점의 색상을 의미
    alpha = 0.5)          # alpha값으로 점의 투명도 조절

# 제목 지정
axes1.set_title('Total Bill vs Tip Colored bySex and Sized by Size')
axes1.set_xlabel('total Bill')
axes1.set_ylabel('Tip')
```

# 04-3 seaborn 라이브러리 자유자재로 사용하기

```python
import seaborn as sns
tips = sns.load_dataset("tips")
```

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'])
ax.set_title('Total Bill Histogram with Density Plot') #경고메시지는 메소드의 사용 방법이 바뀌었음을 알려주는 것.
```

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], kde=False) # kde 값을 False로 지정함에 따라 밀집도 그래프 제외
ax.set_title('Total Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], hist=False) # hist 값을 False로 지정함에 따라 히스토그램 그래프 제외
ax.set_title('Total Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], rug=True)
ax.set_title('Total Bill Histogram with Density and Rug plot')
ax.set_xlabel('Total Bill')
```

#### count 그래프 나타내기

```python
ax = plt.subplots()
ax = sns.countplot('day',data=tips)
ax.set_title('Count of days')
ax.set_xlabel('Day of the week')
ax.set_ylabel('Frequency')
```

#### seaborn 라이브러리로 산점도 그래프 그리기

```python
ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips) #,fig_reg = False 지정 시 회귀선 제거
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

#### 산점도 그래프와 히스토그램을 한번에 그려주는 jointplot 메서드

```python
joint = sns.jointplot(x='total_bill', y='tip', data=tips)
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10,y=1.03)
```

#### 육각그래프(hexbin) 사용, 점이 겹쳐보일 경우 구분하기 어렵다는 문제 해결

```python
hexbin = sns.jointplot(x="total_bill",y="tip", data=tips, kind="hex")
hexbin.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
hexbin.fig.suptitle('Hexbin Joint Plot of Total Bill and Tip', fontsize = 10, y = 1.03)
```

#### 이차원 밀집도 그리기

```python
ax = plt.subplots()
ax =sns.kdeplot(data=tips['total_bill'],
                data2=tips['tip'],
                shade=True)             # shade 인자값을 True로 지정하면 그래프에 음영효과
ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

# 04-5 seaborn 라이브러리로 그래프 스타일 설정하기
- seaborn 라이브러리 스타일은 darkgrid, whitegrid, dark, white, ticks로 총 5 종류입니다.

#### 그래프에 스타일 적용하기

```python
fig, ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True)
```

```python
sns.set_style('darkgrid')
fig, ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True)
```

for문을 이용해 모든 스타일을 다 적용시킨 그래프

```python
fig = plt.figure()
seaborn_styles = ['darkgrid', 'whitegrid','dark','white', 'ticks']
for idx, style in enumerate(seaborn_styles):
    plot_position = idx + 1
    with sns.axes_style(style):
        ax = fig.add_subplot(2,3, plot_position)
        violin = sns.violinplot(x='time', y='total_bill', data=tips, ax=ax)
        violin.set_title(style)

fig.tight_layout()
```
