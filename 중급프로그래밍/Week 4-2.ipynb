#응용문제1 : 임금계산 프로그램

######사용자에게 주당 근무시간과 시간당 임금을 물어본다. 주당 근무시간이 40시간을 넘으면 초과근무시간은 1.5배 임금을 지급해야 한다. 
######이번 주에 받는 총 임금을 계산하는 프로그램을 작성하시오.


```python
worktime = int(input('주당 근무시간은?:'))
wage = int(input('시급은?:'))

if worktime >= 40:
  overtimewage = 1.5
else :
  overtimewage = 1

totalwage = worktime * wage * overtimewage

print(f'이번주 총 임금은 {totalwage}원입니다.')
```

#응용문제2 : 주차료 계산 프로그램
######A학교는 외부인들에게 아래와 같은 주차요금을 부여하고 있다. 
######주차시간을 입력 받아 주차요금을 계산해주는 프로그램을 작성해보세요.

```python
parkingtime = int(input('주차한지 몇분이 지났나요?:'))

if parkingtime > 60:
  parkingfee = 5000 + ((parkingtime-60)//15)*1000
elif parkingtime > 30:
  parkingfee = 5000
else :
  parkingfee = 0


print(f'주차요금은 {parkingfee}원입니다.')
```

#응용문제3 : 자동판매기 프로그램
######자동판매기는 사용자로부터 투입한 돈을 입력 받는다. 물건 값은 100원 단위라고 가정한다. 프로그램은 잔돈을 계산하여 출력한다.
- 음료와 판매금액은 딕셔너리 구조를 사용한다.
- 자판기는 동전 500원, 100원, 50원짜리만 가지고 있다.

```python
beverage = {'coke': 500, 'coffee': 600, 'cider': 500, 'water': 400 }

insertcoin = int(input('돈을 넣어주세요.: '))
choice = int(input('번호를 눌러주세요. : '))


if choice >= 4:
  price = beverage['water']
elif choice >= 3:
  price = beverage['cider']
elif choice >= 2:
  price = beverage['coffee']
else :
  price = beverage['coke']

a = '-'*50
b = '='*50

change = insertcoin - price

print(a)
print('1: coke  500')
print('2: coffee  600')
print('3: cider  500')
print('4: water 400')
print(a)
print(f'투입한 돈: {insertcoin}')
print(f'번호 선택: {choice}')
print(f'거스름돈: {change}')
print(b)

coin500 = change // 500 
change = change % 500 
coin100 = change // 100     
change = change % 100
coin50 = change // 50 
change = change % 50

print(f'500원 동전: {coin500}개')
print(f'100원 동전: {coin100}개')
print(f' 50원 동전: {coin50}개')
print(f'바꾸지 못한 돈: {change}원')
print(b)
```

#응용문제4 : 지불금액 계산 프로그램
######다음은 구매 금액에 따른 할인율을 적용하여 지불 금액을 계산하는 프로그램이다. 단, 금액에 따른 할인율은 아래 표와 같다.

```python
purchase = int(input('구매하신 금액을 알려주세요:'))

if purchase >= 200000:
  discountrate = 0.2
elif purchase >= 50000:
  discountrate = 0.1
elif purchase >= 10000:
  discountrate = 0.05
else :
  discountrate = 0

totalprice = purchase*(1-discountrate)
discountprice = purchase*discountrate
a = '-'*50

print(a)
print(f'구매금액: {purchase}')
print(f'할 인 율: {discountrate*100}%')
print(f'할인금액: {discountprice}')
print(f'지불금액: {totalprice}')
print(a)
```

#응용문제5 : 배송비 계산 프로그램
######배송거리를 입력 받은 후 다음의 조건에 맞는 배송비를 알려주는 프로그램을 작성하시오. 
######단, 00택배회사는 배송거리가 5km 미만일 경우 서비스를 제공하지 않습니다.

```python
deliverydistance = int(input('배송거리를 km 단위로 입력: '))

if deliverydistance >= 500:
  deliveryfee = 6000
elif deliverydistance >= 300:
  deliveryfee = 5000
elif deliverydistance >= 100:
  deliveryfee = 4500
elif deliverydistance >= 50:
  deliveryfee = 4000
elif deliverydistance >= 5:
  deliveryfee = 3500
else :
  deliveryfee = 0

if deliveryfee > 0:
  print(f'{deliverydistance}km에 해당하는 배송비는 {deliveryfee}원입니다.')
else :
  print('5km미만은 배송하지 않습니다.')
```
