##for문 + 리스트 활용1

* 학생 5명의 성적이 리스트에 저장, 합계와 평균을 출력해보시오.



```python
score = [80,50,75,60,90]
sum = 0

#합계
for i in score:
  sum += i

avg = sum/len(score)
print(sum)
print(avg)
```

##for문 + 리스트 활용2

- 학생 5명의 성적이 리스트에 저장, 60점 이상이면 합격, 그렇지 않으면 불합격

```python
score = [80,50,75,60,90]
a = 0
for i in score:
  a += 1
  if i >=60:
    result = "합격입니다."
  else:
    result = "불합격입니다."
  print(f'{a}번학생의 {i}점은 {result}')
  # print(f'{score.index(i) + 1}번 학생은 {result}') index 활용해서 해결도 가능.
```

```python
score = [80,50,75,60,90]
for i in range(len(score)):
  if score[i] >= 60:
    result = "합격입니다."
  else:
    result = "불합격입니다."
  print(f'{i+1}번째 학생의 {score[i]}점은 {result}')
```

```python
# 80점 이상인 학생만 "축하합니다." 메세지를 출력
score = [80,50,75,60,90]
for i in range(len(score)):
  if score[i] < 80:
    continue
  print(f'{i+1} 번째학생 축하합니다.')
```

```python
# 성적합계가 200을 초과할때 까지만 더한 결과를 출력
#실행결과  : 3번째 학생까지의 합계가 205

score = [80,50,75,60,90]
total = 0
for i in range(len(score)):
  total += score[i]
  if total > 200:
    break
print(total)
```

 mlist=['뜨아','아아','라떼','모카'] plist=[2000,3000,3500,4000]

# 응용문제 2 : 스마트폰 연락처 관리 프로그램 만들기

```python
addr = {}

while True:
  s = int(input('1)친구추가 2)친구검색 3)종료 : '))
  if s == 1:
    name = input("이름을 입력하세요. : ")
    if name in addr.keys():  # .keys() 는 없어도됨.
      print("already exist")
    else :
      num = input("번호를 입력하세요 :")
      addr[name] = num
      print("추가되었습니다.")
  elif s == 2:
    search = input("검색할 이름을 입력하세요 : ")
    if search in addr:
      print(search, addr[search])
    else:
      print("not exist")
  elif s == 3:
    break
  else :
    print("잘못된 선택 메뉴입니다.")
```

#응용문제 2-1: 심화문제

```python
addr = {}

while True:
  s = int(input('1)친구추가 2)수정 3)삭제 4)검색 5)전체검색 0)종료 : '))
  if s == 1:
    name = input("이름을 입력하세요. : ")
    if name in addr.keys():  # .keys() 는 없어도됨.
      print("already exist")
    else :
      num = input("번호를 입력하세요 :")
      addr[name] = num
      print("추가되었습니다.")
  elif s == 2:
    name = input("이름을 입력하세요. :")
    if name in addr:
      num = input("변경할 번호를 입력하세요 :")
      addr[name] = num
    else :
      print("연락처에 없는 이름입니다.")
  elif s == 3:
    name = input("이름을 입력하세요.: ")
    if name in addr:
      addr.pop(name)
      print("연락처가 삭제되었습니다.")
    else :
      print("연락처에 없는 이름입니다.")
  elif s == 4:
    search = input("검색할 이름을 입력하세요 : ")
    if search in addr:
      print(search, addr[search])
    else:
      print("not exist")
  elif s == 5:
    for name, number in list(addr.items()):
      print(f'{name}의 전화번호는 {number}입니다.')
  elif s == 0:
    break
  else :
    print("잘못된 선택 메뉴입니다.")
```
