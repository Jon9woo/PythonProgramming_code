# 넘파이 연산
- 전통적인 배열은 동일한 자료형을 가진 데이터를 연속적으로 저장한다.
- 넘파일 배열은 C언어 기반한 배열 구조로 메모리를 적게 차지하고 속도가 빠르다.
- 넘파이 배열을 사용하면 배열 간에 수학적인 연산을 적용
- 고급 연산자와 풍부한 함수를 제공한다.

```python
mid=[10,20,30]
final=[10,20,30]
```

```python
# 리스트
n = len(mid)                      # numpy 사용하면 이러한 복잡한 덧셈코드를 간편하게 사용 가능.
total, avg = [],[]

for i in range(n):
  total.append(mid[i]+final[i])
  avg.append(total[i]/2)

print(total)
print(avg)
```

```python
# 넘파이 배열
mid = np.array([10,20,30])
final = np.array([10,20,30])
total = mid+final
avg = total/2
print(total)
print(avg)
```

```python
# 넘파이 산술연산
a=np.arange(1,11)
b=np.arange(1,11)
print(a+b)
print(np.add(a,b))      #덧셈
print(a-b)
print(np.subtract(a,b)) #뺄셈
print(a*b)
print(np.multiply(a,b)) #곱셈
print(a/b)
print(np.divide(a,b))   #나눗셈
```

```python
# 벡터 수행시간
import time
a = np.arange(10000000)
result = 0

s=time.time()
for v in a:
  result += v
e=time.time()
print(result, e-s) #연산속도 느림

s=time.time()
result = np.sum(a)
e=time.time()
print(result, e-s) #연산속도 빠름
```

# 실습문제 1

```python
import numpy as np
salary=[200,150,180,220,250]
new_salary=np.array(salary)*1.5
print('인상된금액:',new_salary)
avg_salary= new_salary.mean()
print('평균 금액:',avg_salary)
result= (new_salary>=avg_salary)
print('평균이상인 원소:',new_salary[result])
print('평균 이상인 원소의 갯수:',new_salary[result].size)
```
