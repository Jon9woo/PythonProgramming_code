```python
# 1번과제 : 간단한 채팅로봇만들기
import time
print('안녕하세요.')
time.sleep(0.7)
print('나는 인공지능 챗봇입니다.')
time.sleep(0.7)
name = input('당신의 이름은? :')
print(name,'학생 반갑습니다.')
time.sleep(0.7)
major = input('{0}의 전공은 무엇인가요? :'.format(name))
print(name,'학생은 아주 어려운',major,'을 공부하고 계시네요.')
time.sleep(0.7)
print('다음주에 또 만나요.')
```

```python
# 2번과제 : 예산금액 계산하기
attendees = int(input('참석자는 몇명인가요? :'))
chiken = attendees
print('필요한 치킨은', chiken,'마리입니다.')
beer =  attendees*2
print('필요한 맥주는', beer, '캔입니다.')
coke = attendees*2
print('필요한 콜라는', coke, '캔입니다.')

# 치킨가격 = 18000, 맥주가격 = 3000, 콜라가격 = 2000 이라 가정
chikenprice = chiken*18000
beerprice = beer*3000
cokeprice = coke*2000

cokeprice = coke*2000
totalprice = chikenprice + beerprice + cokeprice
print('총 주문금액은', totalprice, '원 입니다.')
```
