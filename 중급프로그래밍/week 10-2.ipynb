#실습예제1 : 은행계좌(BankAccount)

```python
class BankAccount:
  # 생성자
  def __init__(self):
    self.balance=0
  # 출금메소드
  def withdraw(self, amount):
    if amount <= self.balance:
      self.balance -= amount
      print(f'통장에서 {amount}원 출금되었습니다.')
    else:
      print('잔액부족')
  # 입금메소드
  def deposit(self, amount):
    self.balance += amount
    print(f'통장에 {amount}원 입금되었습니다.')
  #
  def __str__(self):
    return f'현재잔액:{self.balance}'
```

```python
# 테스트
b1=BankAccount()
b1.deposit(10000)
print(b1)  
```

# 상속 구현

```python
class Car:
  def __init__(self,model,color,price):
    self.model=model
    self.color=color
    self.price=price 
  def __str__(self):
    result = f'model:{self.model}, color:{self.color}, price:{self.price}' 
    return result
  #일반메소드
  def setColor(self, color):
    self.color=color
  def getColor(self):
    return self.color
```

```python
class ElectricCar(Car):
  def __init__(self, model, color, price, batterySize):
    super().__init__(model,color,price)   # 생성자는 명시적으로 호출해줘야함.
    self.batterySize = batterySize
  def __str__(self):
    result = super().__str__()
    result += f'batterySize:{self.batterySize}'
    return result
```

```python
# 자식객체생성
e1=ElectricCar('Model5','red','1000',0)
print(e1)
e1.model  #부모에 정의된 변수들은 접근 가능 but 부모의 생성자는 상속되지 않음!! (super키워드 필요)

e1.setColor('yellow')
print(e1.getColor())
```

# 도전과제 : 성적 관리 프로그램

```python
class ProgramLecture:
  #생성자
  def __init__(self):
    self.assignment = 0
    self.firstExam = 0
    self.secondExam = 0
    self.absence = 0

  #입력데이터
  def inputData(self,assignment, firstExam, secondExam, absence):
    self.assignment = assignment
    self.firstExam = firstExam
    self.secondExam = secondExam
    self.absence = absence

  #평균점수
  def getAvg(self):
    avg = (self.assignment+self.firstExam+self.secondExam)/3
    avg = round(avg, 2)
    return avg

  #가중치평균점수
  def getWeightAvg(self):
    weight_1 = 20
    weight_2 = 30
    weight_3 = 50
    weightAvg = (self.assignment*weight_1+self.firstExam*weight_2+self.secondExam*weight_3)/100
    return weightAvg

  #학점
  def getGrade(self):
    result = self.getWeightAvg()
    if self.absence >= 4:
      grade = '출석 미달 Fail'
    elif result >= 90:
      grade = 'A'
    elif result >=80:
      grade = 'B'
    elif result >= 70:
      grade = 'C'
    elif result >= 60:
      grade = 'D'
    else:
      grade = 'F'
    return grade

  #
  def __str__(self):
    result = f'최종 점수는 {self.getAvg()} 학점은 {self.getGrade()}입니다.'
    return result
  


#메인코드

## 데이터 입력
gradeInput = '=====성적입력====='
print(gradeInput)

data1 = input('1번학생:')
data2 = input('2번학생:')
data3 = input('3번학생:')
studentdata = [data1, data2, data3]
finaldata=[]

for i in range(len(studentdata)):
  temp = []
  data = studentdata[i].split(',')
  temp.append(int(data[0]))
  temp.append(int(data[1]))
  temp.append(int(data[2]))
  temp.append(int(data[3]))
  finaldata.append(temp)

## 데이터 출력
gradeOutput = '=====성적산출====='
print(gradeOutput)

pl = ProgramLecture()
for i in range(len(studentdata)):
 pl.inputData(finaldata[i][0],finaldata[i][1],finaldata[i][2],finaldata[i][3])
 print(pl)
```

# 도전(응용)문제 2 : 회원가입 프로그램(dic)
 - 딕셔너리 구조 사용

```python
class Member:
  def __init__(self,id,pw):
    self.mId=id
    self.mPw=pw

  def __str__(self):
    return f'{self.mId}, {self.mPw}'

class MemberManage:
  def __init__(self):
    self.members={}

  def addMember(self,m):
    self.members[m.mId] = m.mPw

  def loginMember(self,id,pw):
    if id in self.members:
      if self.members[id] == pw:
        print('로그인 성공')
      else:
        print('비밀번호가 다릅니다.')
    else:
      print('해당 ID가 없습니다.')

  def removeMember(self,id):
    if id in self.members:
      del self.members[id]
      print('아이디가 삭제되었습니다.')

  def printMember(self):
    print('---전체회원---')
    for m in self.members:
      print(f'ID : {m}')
      print(f'PW : {self.members[m]}')
      print('-'*20)
  
  def __str__(self):
    result = super().__str__()
    return result
```

#도전(응용)문제 2: 회원가입 프로그램(list)
 - 리스트 구조 사용

```python
class Member:
  def __init__(self,id,pw):
    self.mId=id
    self.mPw=pw

  def __str__(self):
    return f'{self.mId}, {self.mPw}'

class MemberManage:
  def __init__(self):
    self.members=[]

  def addMember(self,m):
    self.members.append([m.mId, m.mPw])  #튜플로 넣는게 더 좋음. (비밀번호는 변경되면 안되니까..?) #리스트에는 다른 자료형을 넣을 수 있다.

  def loginMember(self,id,pw):
    count = 0
    for i in range(len(self.members)):
      if self.members[i][0] == id:
        if self.members[i][1] == pw:
         print('로그인 성공')
         break
        else:
         print('비밀번호가 다릅니다.')
      else:
        count += 1
        if count == len(self.members):
          print("해당 ID가 없습니다.")

  def removeMember(self,id):
    for i in range(len(self.members)):
      if self.members[i][0] == id:
         del self.members[i]
         print('아이디가 삭제되었습니다.')
         break
      else:
        print('해당 ID가 없습니다.')


  def printMember(self):
    print('---전체회원---')
    for i in range(len(self.members)):
      print(f'ID : {self.members[i][0]}')
      print(f'PW : {self.members[i][1]}')
      print('-'*20)
  
  def __str__(self):
    result = super().__str__()
    return result
```

```python
# 메인코드
m=MemberManage()
m.addMember(Member('hee8460','1234'))
m.addMember(Member('hee8461','1234'))
m.addMember(Member('hee8462','1234'))

# 전체조회
m.printMember()

# 로그인 인증처리
m.loginMember('hee8460','1234')
m.loginMember('hee8465','1234')

# 회원삭제
m.removeMember('hee8460')
```

```python
m.printMember()
```

# TV클래스 정의문제

```python
class Tv:
  def __init__(self):
    self.channel=1
    self.volume=0
    self.power=False

  def turnOn(self):
    self.power = True

  def turnOff(self):
    self.power = False

  def setChannel(self,channel):
    self.channel = channel
  #설정자(set~)메소드 / 접근자(get~)메소드
  def setVolume(self,volume):
    self.volume = volume
  
  def getVolume(self):
    return self.volume


  def __str__(self):
    if self.power == False:
      return("응답이 없습니다.")
    else:
      return f'TV의 채널:{self.channel}, TV의 볼륨: {self.volume}'
```

```python
tv = Tv()
```

```python
tv.turnOn()
```

```python
tv.setChannel(11)
tv.setVolume(6)
```

```python
print(tv)
```
