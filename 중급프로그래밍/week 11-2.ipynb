#메소드 오버라이딩
: 부모 클래스의 메소드를 자식 클래스가 자신의 필요에 맞추어서 변경

```python
class Bank():
  def getInterestRate(Self):
   return 0.0     # 의미없는 0 설정, pass 처리할 수도 o

class BadBank(Bank):
  def getInterestRate(self):   #메소드 오버라이딩
    print(super().getInterestRate()) #명시적으로 부모의 멤버메소드를 호출하는 함수
    return 10.0

class NormalBank(Bank):
  def getInterestRate(self):   #메소드 오버라이딩
    return 5.0

class GoodBank(Bank):
  def getInterestRate(self):   #메소드 오버라이딩
    return 3.0          



b1 = BadBank()
b2 = NormalBank()
b3 = GoodBank()
print(f'BadBank의 이자율: {b1.getInterestRate()}')
print(f'NormalBank의 이자율: {b2.getInterestRate()}')
print(f'GoodBank의 이자율: {b3.getInterestRate()}')
```

# 상속의 구현 예제
- 사람 클래스 : 이름, 나이
- 학생 클래스 : 이름, 나이, 학과
## 상속의 장점
- 클래스 사이의 멤버 중복 선언 불필요
- 필드(멤버변수)와 메소드 재사용으로 클래스 간결화
- 클래스 간 계층적 분류 및 관리

```python
#부모클래스 = 슈퍼클래스 = 기저(base)클래스 = 상위클래스
class Person():
  # 멤버변수  
  name=''
  age=0                       #얘네는 계속 초기화 + Person('a','b') 이런식으로 안에 사용
  # 생성자
  def __init__(self,name,age):
    self.name=name
    self.age=age
  # 일반메소드
  def setName(self,name):
    self.name=name
  def setAge(self,age):
    self.age=age
  def work(self):
    print('일은 한다')
  # 출력메소드
  def show(self):
    print(f'{self.name}의 나이는 {self.age}입니다.')
```

```python
#자식클래스 = 하위클래스 = 유도클래스
class Student(Person):
  def __init__(self,name,age,dept):
    super().__init__(name,age)
    self.dept=dept
  def work(self):
    super().work()
    print('공부를 한다')
  def show(self):
    super().show()
    print(f'학과는 {self.dept} 입니다.')
```

```python
# 테스트
p1 = Person('이종우',20)
p1.show()
```

```python
s1=Student('이종우',26,'DA')
s1.show()
s1.setName('이종우야호')
s1.show()
s1.work()
```

#실습예제2: 직원과 매니저
회사에 직원(Employee)과 매니저(Manager)가 있다.
직원은 월급만 있지만 매니저는 월급 외에 보너스가 있다고 하자.
- Employee 클래스를 상속받아서 Manager 클래스를 작성한다.
- Employee 클래스의 getSalary()는 Manager 클래스에서 재정의된다.

```python
class Employee:
  def __init__(self,name,salary):
    self.name=name
    self.salary=salary

  def getSalary(self):
    return self.salary

  def __str__(self):
    result=f'이름:{self.name}, 월급:{self.salary}'
    return result

class Manager(Employee):
  def __init__(self,name,salary,bonus):
    super().__init__(name,salary)
    self.bonus=bonus

  def getSalary(self):
    salary = super().getSalary()
    salary=salary+self.bonus
    return salary

  def __str__(self):
    print(super().__str__(), end=' ')
    result = f'보너스:{self.bonus}'
    return result
```

```python
e1 = Employee('이종우', 10000)
print(e1)
m1 = Manager('이종우', 10000, 3000)
print(m1.getSalary())
print(m1)
```

# 도전문제 : Person 클래스
- Person 클래스를 기반으로 Student와 Teacher이라는 두 개의 자식 클래스를 구현한다.

```python
# 부모클래스
class Person():
  def __init__(self,name,number):
    self.name = name
    self.number = number

# 자식클래스_Student
class Student(Person):
  UNDERGRADUATE = 0
  POSTGRADUATE = 1

  def __init__(self,name, number,studentType):
    super().__init__(name,number)
    self.studentType = studentType
    self.classes = []
    self.gpa = 0

  def enrollCourse(self,subject):
    self.classes.append(subject)

  def __str__(self):
    result = f'이름:{self.name}, \n주민번호:{self.number}, \n수강과목:{self.classes}, \n평점:{self.gpa} '
    return result

# 자식클래스_Teacher
class Teacher(Person):

  def __init__(self,name,number):
    super().__init__(name,number)
    self.courses = []
    self.salary = 3000000

  def assignTeaching(self,subject):
    self.courses.append(subject)
    
  def __str__(self):
    result = f'이름:{self.name}, \n주민번호:{self.number}, \n강의과목:{self.courses}, \n월급:{self.salary} '
    return result
```

```python
#학생 2명 
hong=Student('홍길동','12345678', Student.UNDERGRADUATE)
hong.enrollCourse('중급프로그래밍')
hong.enrollCourse('파이썬')
hong.enrollCourse('빅데이터분석')


lee=Student('이길동','13245768', Student.UNDERGRADUATE)
lee.enrollCourse('텍스트마이닝')
lee.enrollCourse('JAVA')
lee.enrollCourse('딥러닝')
```

```python
print(hong)
print(lee)
```

```python
#강사 2명
kim=Teacher('김철수','123456790')
kim.assignTeaching('Python')
kim.assignTeaching('머신러닝')

park=Teacher('박철수','13245960')
park.assignTeaching('C++')
park.assignTeaching('추천시스템')
```

#도전문제2 : Ticket 클래스

```python
#부모 클래스
class Ticket:
  def __init__(self,number,price):
    self.number = number
    self.price = price

  def getNumber(self):
    print(self.price)

  def getPrice(self):
    return self.price

  def setPrice(self,price):      # set price를 할거면 처음 __init__에서 지정안해줘도 됌.
    self.price = price

  def __str__(self):
    result = f'입장권번호:{self.number}, \n입장권가격:{self.price}'
    return result

#자식클래스1
class WalkupTicket(Ticket):  #일반입장권
  creditcard = True
  cash = False

  def __init__(self,number,price,payBYcredit):
    super().__init__(number,price)
    self.payBYcredit = payBYcredit

  def getPrice(self):                      #메소드 오버라이딩
    if self.payBYcredit == True:
      extraChargeRate = 1.05
    else:
      extraChargeRate = 1
    self.price = super().getPrice() * extraChargeRate
    return self.price

  def getPayByCredit(self):
    if self.payBYcredit == True:
      result = "신용카드 결제입니다."
    else:
      result = "현금 결제입니다."
    return result

  def __str__(self):
    result = f'입장권번호:{self.number}, 입장권가격:{int(self.getPrice())}, 신용카드결제:{self.payBYcredit}'
    return result


#자식클래스2
class AdvanceTicket(Ticket):  #일반입장권

  def __init__(self,number,price,advancedays):
    super().__init__(number,price)
    self.advancedays = advancedays

  def getPrice(self):                      #메소드 오버라이딩
    if self.advancedays >= 30:
      discountRate = 0.5
    elif self.advancedays >= 10:
      discountRate = 0.2
    elif self.advancedays >= 5:
      discountRate = 0.1
    else:
      discountRate = 0
    self.price = super().getPrice() - (super().getPrice() * discountRate)
    return self.price

  def getAdvancedays(self):
    return self.advancedays

  def __str__(self):
    result = f'입장권번호:{self.number}, 입장권가격:{int(self.getPrice())}, 사전예약일수:{self.advancedays}'
    return result
```
