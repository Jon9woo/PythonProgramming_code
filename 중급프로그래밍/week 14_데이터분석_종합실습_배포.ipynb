# 데이터 시각화

## 데이터 불러오기

```python
import pandas as pd
import matplotlib.pyplot as plt
filename='/content/drive/MyDrive/data/fine_dust.xlsx'
data = pd.read_excel(filename, index_col='area')
data2019=data[2019]
data
```

```python
data[2019]
```

## 꺽은선 그래프

```python
# 2016-2019 꺾은 선 그래프 그리기
plt.figure(figsize=(20,5))
year = [2016,2017,2018,2019]
for i in year:
  plt.plot(data[i],color='g',marker='o')
plt.xlabel('area')   # x축 이름
plt.ylabel('micrometer')  # y축 이름
plt.show()
```

## 막대 그래프

```python
data2016 = data[2016]
plt.figure(figsize=(15,4))
plt.bar(data2016.index, data2016, color='g')
plt.ylim(35,55)
plt.xlabel('area')
plt.ylabel('micrometer')
plt.title('2016 Fine Dust Bar Graph')
plt.grid()
plt.show()
```

# 데이터 전처리
- 결측데이터 처리하기

## 결측치 확인하기

```python
titanic.isnull().sum()
```

## 결측치가 많은 컬럼 삭제

```python
# 객실번호(Cabin) 컬럼 삭제
titanic.drop(['Cabin'],axis=1,inplace=True)
titanic
```

```python
# 컬럼확인
titanic.columns
```

## 결측데이터 최빈값으로 대체하기

```python
# 탑승항구(Embarked)는 최빈값으로 대체
# 최빈값 찾기
titanic['Embarked'].value_counts()
```

```python
# 최빈값으로 대체하기
titanic['Embarked'].fillna('S',inplace=True)
```

```python
titanic.isnull().sum()
```

```python
# 나이(Age)는 평균구하기
avg=titanic['Age'].mean()
avg
```

```python
# 결측데이터 평균값으로 대체하기
titanic['Age'].fillna(avg,inplace=True)
titanic.isnull().sum()
```

```python
# 결측데이터 확인하기
titanic.isnull().sum()
```

# 데이터 그룹핑하기

```python
# 1. 항구(Embarked)별 평균값을 출력
titanic.groupby('Embarked').mean()
```

# 판다스의 시각화

## 생존자 확인
시본모듈로 막대그래프 그리기

```python
# 좌석등급(Pclass)별 생존자(Survived) 확인
sns.countplot('Pclass',hue='Survived',data=titanic)
```

```python
# 성별(Sex)별 생존자(Survived) 확인
sns.countplot('Sex',hue='Survived',data=titanic)
```

```python
# 20세 미만의 연령대별 생존/사망 여부 확인
df=titanic['Age']<20
underage20=titanic[df]
underage20
```

```python
plt.figure(figsize=(15,5))
sns.countplot('Age',hue='Survived',data=underage20)
```

## 상관관계확인 
- 시본 모듈로 히트맵 그리기

```python
# 상관관계지수
corr=titanic.corr()
corr
sns.heatmap(data=corr,annot=True,cmap='Blues')
```
