# 얕은복사 vs 깊은복사

```python
#변경 가능한 객체
arr1 = [10,20,30]
arr2 = arr1 #주소를 복사하는 것.
#arr2[0] = 100
arr2.append(40)
arr1, arr2
```

```python
#변경 불가능한 객체
a1 = 10
a2 = a1
print(id(a1), id(a2))
a2 = 100 #새로운 오브젝트가 할당이 되면 a2의 주소는 변경됨.
print(id(a1), id(a2))
```

```python
#리스트
a=[1,2,3,4]
b=a.copy() # a[:]
print(id(a), id(b))
print(id(a[0]), id(b[0]))
b[0] = 10
print(a,b)
print(id(a[0]), id(b[0]))
```

```python
#리스트 내 리스트 : 이중리스트
a = [[1,2],[3,4]]
b=a.copy() 
print(id(a), id(b))
print(id(a[0]), id(b[0]))
b[0].append(30)
a,b
```

```python
# 깊은복사(deepcopy)
import copy
a = [[1,2],[3,4]]
b=copy.deepcopy(a)
print(id(a),id(b))
print(id(a[0]),id(b[0]))
b[0].append(30)
a,b
```
