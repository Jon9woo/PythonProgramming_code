```python
# 예금인출기능과 출금기능을 함수로 정의해보세요.
# 잔액변수(전역변수)
def deposit(amount):
  global balance
  balance += amount
  print(f'입금 후 잔고는 {balance} 입니다')

def withdraw(amount):
  global balance
  if amount > balance :
    print('잔고가 부족합니다.')
  else:
    balance -= amount
    print(f'출금 후 잔고는 {balance} 입니다')
    

balance = 0

while True:
  s = int(input('1)입금기능 2)출금기능 3)종료: '))
  if s == 1:
    amount = int(input("금액을 입력하세요."))
    deposit(amount)
  elif s == 2 :
    amount = int(input("금액을 입력하세요."))
    withdraw(amount)
  elif s == 3 :
    break
  else:
    '잘못눌렀습니다.'
```

# 응용문제1 : 문자열 반복 (완료)
- 문자열과 반복횟수를 매개변수로 전달 받으면 반복횟수 만큼 문자열을 반복하여 출력하는 프로그램을 작성해보세요.

```python
# 함수정의
def repeat(x, y):
  a = 0
  while True:
    if a < y: 
      a += 1
      print(x)
    else :
      break

# 메인코드
sentence = input('문자열 :')
times = int(input('횟수 :'))
repeat(sentence, times)
```

#응용문제 2 : 계산기 프로그램 (완료)
- 사용자로부터  두  정수를입력  받아  사칙  연산  결과를  출력해보세요.

```python
# 함수정의
def plus(x,y):
  print(x+y)

def minus(x,y):
  print(x-y) 

def multi(x,y):
  print(x*y)

def divid(x,y):
  if y == 0 :
    print('0으로는 나눌 수 없습니다.')
  else :
    print(x/y)

# 메인코드
num1 = int(input('첫번째 정수는? :'))
num2 = int(input('두번째 정수는? :'))

while True:
  s = int(input('1:add 2:sub 3:mul 4:div =>' ))
  if s == 1:
    plus(num1, num2)
  elif s == 2 :
    minus(num1,num2)
  elif s == 3:
    multi(num1,num2)
  elif s == 4:
    divid(num1,num2)
  elif s == 0:
    break
  else:
    print('잘못 입력하였습니다.')
```

# 응용문제 4 : 스마트폰 연락처 관리 프로그램  (완료)
- 5주차 스마트폰 프로그램에서   1) 추가기능 2)수정기능 3) 삭제기능 4) 검색기능 5) 전체검색기능을 각각 함수로 정의한 후 호출해보세요.

```python
addr = {}

#추가기능 함수
def add():
    name = input("이름을 입력하세요. : ")
    if name in addr.keys(): 
      print("already exist")
    else :
      num = input("번호를 입력하세요 :")
      addr[name] = num
      print("추가되었습니다.")
#수정기능 함수
def revise():
    name = input("이름을 입력하세요. :")
    if name in addr:
      num = input("변경할 번호를 입력하세요 :")
      addr[name] = num
    else :
      print("연락처에 없는 이름입니다.")
#삭제기능 함수
def delete(x):
  if x in addr:
      addr.pop(x)
      print("연락처가 삭제되었습니다.")
  else :
      print("연락처에 없는 이름입니다.")
#검색기능 함수
def search(x):
    if x in addr:
      print(x, addr[x])
    else:
      print("not exist")
#전체검색기능 함수
def searchall():
  for name, number in list(addr.items()):
    print(f'{name}의 전화번호는 {number}입니다.')


while True:
  s = int(input('1)친구추가 2)수정 3)삭제 4)검색 5)전체검색 0)종료 : '))
  if s == 1:
    add()
  elif s == 2:
    revise()
  elif s == 3:
    x = input("이름을 입력하세요.: ") # 위에 2개 함수처럼 이 과정도 함수에 넣는게 더 좋을까요? (질문)
    delete(x)
  elif s == 4: 
    x = input("검색할 이름을 입력하세요 : ")
    search(x)
  elif s == 5:
    searchall()
  elif s == 0:
     break
  else :
    print("잘못된 선택 메뉴입니다.")
```
