#응용문제5

```python
training_data = [['노랑', 2,'개나리'],
                 ['흰색', 5, '벚꽃'],
                 ['노랑', 5, '개나리'],
                 ['흰색', 2, '목련'],
                 ['분홍', 2, '진달래']]
```

```python
def flower_counts(data):
  flower_counts_dict = {}   #딕셔너리 
  for i in data:
    if i[2] not in flower_counts_dict:   # 리스트에서 0 , 1 , 2 니까 세번째 요소
      flower_counts_dict[i[2]] = 1   #처음 count 되면 1로 지정
    else:
      flower_counts_dict[i[2]] += 1  #이후 count 될때마다 +1 씩

  return flower_counts_dict

flower_counts(training_data) 
```
