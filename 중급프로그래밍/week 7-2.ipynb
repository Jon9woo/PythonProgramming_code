 1. 무한 반복문을 이용해 로또 번호 6개를 출력한다.

```python
import random as rd
rd.randint(1,45)
```

```python
#함수정의
def getNum():
  return rd.randint(1,45)

#메인코드
lotto_list = []
while True :
  lotto = getNum()
  if lotto  not in lotto_list:
      lotto_list.append(lotto)
      if len(lotto_list) > 5:
        break
print(f'추첨된 로또번호는 {lotto_list} 입니다.')
```

#자율실습문제

```python
filename = '/content/drive/MyDrive/data/LetitbeLyrics.txt'
f = open(filename)
lyrics = f.read()
split_lyrics = lyrics.split() 


#함수정의
def wordcount (data):
  word_list = {}
  for i in data:
    if i not in word_list:
      word_list[i] = 1
    else :
      word_list[i] += 1
  return word_list

#메인코드
wordcount(split_lyrics)
```

```python
#더 나아가기 : 빈도수가 가장 높은 단어를 찾아보자
def wordcount2 (data):
  word_list = {}
  for i in data:
    if i not in word_list:
      word_list[i] = 1
    else :
      word_list[i] += 1
  max_key = max(word_list, key=word_list.get)
  return max_key

#메인코드
wordcount2(split_lyrics)
```

# 응용문제1 : 성적프로그램

```python
#텍스트파일 읽기
score = '/content/drive/MyDrive/data/score.txt'
f=open(score)
data = f.read().split('\n') 

# 평균 계산 함수정의
def cal_avg(i):
  kor = int(data[i-1].split(',')[1])
  eng = int(data[i-1].split(',')[2])
  math = int(data[i-1].split(',')[3])
  return round((kor+eng+math)/3, 1)   #소숫점 첫 번째 자리까지만

# 학점등급계산기 함수정의
def grade(x):
  if x >= 90:
    result ='A'
  elif x >= 80:
    result ='B'
  elif x >= 70:
    result ='C'
  elif x >= 60:
    result ='D'
  else:
    result ='F'
  return result
  
  
# 메인코드
print('이름  ', '국어  ', '영어  ', '수학  ', '평균  ', '학점  ')
print('-' * 40)
for i in range(len(data)):
  print(data[i].split(',')[0], end='  ')
  print(data[i].split(',')[1], end='     ')
  print(data[i].split(',')[2], end='     ')
  print(data[i].split(',')[3], end='    ')
  avg = cal_avg(i)
  print(avg, end='   ')
  print(grade(avg))

# 김연수 학생 공백 정렬 방법??? 
```

# 응용문제2 :  초 간단 타자게임 만들기

```python
#텍스트파일 읽기
food = '/content/drive/MyDrive/data/word.txt'
f=open(food)
data = f.read().split('\n') 
print(data)

import random as rd
rd.choice(data)

# 게임프로그램 함수정의
def game(data):
  while True:
    a = rd.choice(data)
    print(a)
    s = input()
    if a == s:
      print('맞음!!')
    elif s == '':     # 프로그램 종료는 사용자가 엔터를 입력하면 강제 종료된다.
      break
    else:
      print('오타! 다시 도전')


#메인코드
print('-'*24)
print('내가 먹고싶은 음식 게임')
print('-'*24)
print('타자게임 시작(아무키나)')
game(data) #게임프로그램. () 함수 내부에 입력하는 파일에 따라 다른 단어들이 나옴.
```
