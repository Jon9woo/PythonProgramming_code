세트 : 얘는 집합이라고 이해하면 될듯

```python
set1 = {10, 20, 30}
print(set1)

set1.add(40)
print(set1)

set1.update([50,60])
print(set1)

set2 = {30,40,50}

print(set1.union(set2))
set1.union(set2)

set1.difference(set2) #차집합
set1.intersection(set2) #교집합
```
