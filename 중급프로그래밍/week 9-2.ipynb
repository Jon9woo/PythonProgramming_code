#클래스 정의

```python
class Person:
  #변수
  #함수(메소드)
  def setName(self,name,age): #self변수 = 객체자신
    self.name = name
    self.age = age
  def show(self):
    print(f'name:{self.name} : {self.age}')
```

```python
# 객체 생성
p1 = Person()
p1.setName('나행복',26)
p1.show()

p2 = Person()
p2.setName('너행복',29)
p2.show()
```

```python
#생성자 정의(선언) : 객체의 초기화
class Person:
  #생성자
  def __init__(self,n,a):
    self.name=n
    self.age=a   #n과a가 매개변수
  def show(self):
    print(f'name:{self.name}: {self.age}')
```

```python
#객체 생성
p1 = Person('나행복',22)
p1.show()
```

```python
class Calculator:
  def __init__(self,n1,n2):
    self.num1=n1
    self.num2=n2
  def add(self):
    result=self.num1+self.num2
    return result

  def sub(self):
    result=self.num1-self.num2
    return result
```

```python
#객체 생성
c1 = Calculator(10,20)
c1.add()
```

```python
#객체 속성(멤버변수) 출력(접근)
print(c1.num1)
print(c1.num2)
```

```python
#객체 메소드 접근
print(c1.add())
print(c1.sub())
```

```python
#객체 멤버변수 변경
c1.num1 = 100
c1.num2 = 200
```

```python
print(c1.add())
```

#도전문제1 : 자동차 클래스 정의 및 객체 활용

```python
class Car:
  def __init__(self,color,speed):
    #생성자
    self.color=color
    self.speed=speed
  def upSpeed(self,value):
    self.speed+=value
  def downSpeed(self,value):
    self.speed-=value
  def printinfo(self):
    print(f'color: {self.color}, speed:{self.speed}')
  def __str__(self):
    msg = f'color:{self.color}, speed : {self.speed}'
    return msg
```

```python
#객체생성
c1 = Car('blue',0)
c2 = Car('red', 0)
#객체정보출력
print(f'color: {c1.color}, speed:{c1.speed}')
print(f'color: {c2.color}, speed:{c2.speed}') #이렇게 print 할필요없다

#객체정보 함수로 출력
c1.printinfo() # -> 더 좋은 코드
```

#도전문제2 : 학생 성적 관리 프로그램

```python
# 클래스 선언
class Grade:
  # 생성자
  def __init__(self,name,kor,eng,math):
    self.name = name
    self.kor = kor
    self.eng = eng
    self.math = math
  # 합계기능
  def getSum(self):
    sum = self.kor+self.eng+self.math
    return sum
  # 평균기능
  def getAvg(self):
    avg = round((self.kor+self.eng+self.math)/3, 1)
    return avg
  # 학점기능
  def getGrade(self):
    result = (self.kor+self.eng+self.math)/3
    if result >= 90:
      grade = 'A'
    elif result >=80:
      grade = 'B'
    elif result >= 70:
      grade = 'C'
    elif result >= 60:
      grade = 'D'
    else:
      grade = 'F'
    return grade
  # 전체출력기능
  def printinfo(self):
    print(f'이름:{self.name}, 평균 : {self.getAvg()}, 학점: {self.getGrade()}')
  # 객체설명기능
  def __str__(self):
    msg = f'이름:{self.name}, 평균 : {self.getAvg()}, 학점: {self.getGrade()}'
    return msg

# 파일불러오기
filename = '/content/drive/MyDrive/data/score.txt'
f = open(filename)
data = f.read().splitlines()

# 형 변환 / 리스트화
for i in range(len(data)):
  score=[]
  for j in range(len(data)):
    A = []
    A.append(str(data[j].split(',')[0]))
    A.append(int(data[j].split(',')[1]))
    A.append(int(data[j].split(',')[2]))
    A.append(int(data[j].split(',')[3]))
    score.append(A)

# 메인 코드
line = '=' * 35
print(line)
for i in range(len(score)):
  s1 = Grade(score[i][0], score[i][1], score[i][2], score[i][3])
  s1.printinfo()
print(line)
```
